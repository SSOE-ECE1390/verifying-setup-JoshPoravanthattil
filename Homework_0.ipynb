```python
# python imports for homework assignment
import cv2
import matplotlib.pyplot as plt
```

```python
# 1. Load an image in grayscale
img_path = "Data/Apollo_11_Launch.jpg"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
```

```python
# 2. View image
def show_image(image, title="Image"):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

show_image(img, "Original Image")
```

```python
# 3. Crop image (e.g., center crop)
h, w = img.shape
cropped_img = img[h//6: 5*h//6, 2*w//5: 3*w//5]  # crop middle square
show_image(cropped_img, "Cropped Image")
```

```python
# 4. Edit brightness (increase brightness by +50)
bright_img = cv2.convertScaleAbs(cropped_img, alpha=1, beta=50)  # alpha=contrast, beta=brightness
show_image(bright_img, "Brightness Adjusted")
```

```python
# 5. Rotate image (90 degrees clockwise)
rotated_img = cv2.rotate(bright_img, cv2.ROTATE_90_CLOCKWISE)
show_image(rotated_img, "Rotated Image")
```

```python
# 6. Save modified image
save_path = "Data/img_modified.jpg"
cv2.imwrite(save_path, rotated_img)
print(f"Modified image saved at {save_path}")
```
